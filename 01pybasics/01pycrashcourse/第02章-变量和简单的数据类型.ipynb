## 第2章　变量和简单的数据类型

### 2.1 变量

**变量是存储数据的名称，是可以被赋值的标识符，在程序中可以随时修改变量值**，命名规则如下：
- 变量名只能包含字母、数字和下划线，不能以数字开头
- 变量名不能使用Python关键字、函数名
- 变量名不能包含空格等特殊字符
- 变量名区分大小写，并且尽量使用小写的变量名

变量可以同时多个被赋值，**变量名和数值之间用逗号分隔**，如：

```python
a, b, c = 1, 2, 3
print(a, b, c)
```

### 2.2 字符串

**字符串是被单引号或双引号括起来的字符序列**，常见的操作如下：
- title()：将字符串中的每个单词首字母大写，其他字母小写
- rstrip()：删除字符串末尾的空白字符
- lstrip()：删除字符串开头的空白字符
- strip(): 删除字符串开头和结尾的空白字符
- removeprefix(): 删除字符串开头的指定前缀
- removesuffix(): 删除字符串结尾的指定后缀

**f 字符串**：在字符串中插入变量值的方法，方法如下：

```python
name = "Alice"
age = 18
print(f"我的名字叫{name}，今年{age}岁")
```

### 2.3 数

数包括整数和浮点数，常见的操作如下：
- 数之间可以进行+、-、*、/、**运算
- 整数和浮点数之间可以进行运算，结果为浮点数；任意两个数相除，结果为浮点数
- 大数可以用下划线来分隔，如：123_456_789，输出结果时会去掉下划线

Python中没有内置的常量类型，通常用全大写的变量名来表示常量。

特别注意浮点数的运算结果，包含的小数位数可能是不确定的，如：0.1 + 0.2 != 0.3

```python
0.1 + 0.2
```

### 2.4 注释

在Python中，注释的方法有如下几种：
- 单行注释：使用 # 开头
- 多行注释：使用 ''' 或 """ 开头和结尾
- 文档字符串：使用 ''' 或 """ 开头和结尾
