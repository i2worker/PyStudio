## 第4章 操作列表

### 4.1 遍历列表

- 通过 **for...in...** 循环遍历列表中所有元素。

```python
magicians = ['alice', 'david', 'carolina']
for magician in magicians:
    print(f'{magician.title()}, that was a great trick!')
    print(f'I can\'t wait to see your next trick, {magician.title()}.\n')
print('Thank you, everyone. That was a great magic show!')
```

### 4.2 数值列表

- 通过 **range(start, stop, step)** 函数创建数列，通过 **list()** 函数将 range() 函数返回的数列转换成列表。

```python
numbers = list(range(2, 11, 2))
print(numbers)
```

- 可以对数值列表进行简单的统计计算。

```python
numbers = list(range(1, 10))
numbers.append(0)
print(numbers)
print(min(numbers))
print(max(numbers))
print(sum(numbers))
```

- 列表推导式：将 **for** 循环的结果保存在一个列表中。

```python
squares = [value ** 2 for value in range(1, 11)]
print(squares)
```

### 4.3 切片

- 获取列表中的部分元素称为切片。

```python
players = ['charles', 'martina', 'michael', 'florence', 'eli']
print(players[0:3])
print(players[1:4])
print(players[:4])
print(players[2:])
print(players[-3:])
```

- 复制列表：通过创建一个包含整个列表的切片来复制列表。

```python
foods1 = ['apple', 'pear', 'orange', 'banana']
foods2 = foods1[:]
foods1.append('coconut')
foods2.append('mango')
print(foods1)
print(foods2)
```

### 4.4 元组

- 元组：用 **(,)** 创建，其中的元素不可以修改。

```python
dimensions = (100, 200)
print(dimensions)
print(dimensions[0])
print(dimensions[1])
```
