## 第3章 列表简介

### 3.1 列表是什么

- 列表由一系列按特定顺序排列的元素组成，**其中的元素类型可以不同**，用 [] 来表示，用逗号隔开元素。

```python
bicycles = ['trek', 'cannondale', 'redline', 'specialized']
print(bicycles)
```

- 通过索引访问列表中的元素，索引从 0 开始，也可用 -1 访问列表中最后一个元素。

```python
print(bicycles[0])
print(bicycles[-1])
```

### 3.2 列表操作

- 修改元素：指定列表名和索引，再指定该元素的新值。

```python
motorcycles = ['honda', 'yamaha', 'suzuki']
print(motorcycles)

motorcycles[0] = 'ducati'
print(motorcycles)
```

- 添加元素：**append()** 方法添加到列表末尾，**insert()** 方法添加到指定位置。

```python
motorcycles = ['honda', 'yamaha', 'suzuki']
print(motorcycles)

motorcycles.append('ducati')
print(motorcycles)

motorcycles.insert(0, 'ducati')
print(motorcycles)
```

- 删除元素：**del** 语句删除列表中指定的元素，**pop()** 方法删除列表末尾元素并返回该元素，**remove()** 方法根据值删除第一个匹配的元素(如果有多个相同值则需要通过循环删除)。

```python
motorcycles = ['honda', 'yamaha', 'suzuki']
motorcycles.append('ducati')
motorcycles.insert(0, 'ducati')
print(motorcycles)

del motorcycles[-1]
print(motorcycles)

first = motorcycles.pop(0)
print(motorcycles)
print(first)

motorcycles.remove('honda')
print(motorcycles)
```

- 排序元素：**sort()** 方法对元素进行永久排序(reverse=True可以实现反向排序)，**sorted()** 方法对元素进行临时排序(返回一个新列表)，**reverse()** 方法永久反转列表元素排序。

```python
cars = ['bmw', 'audi', 'toyota', 'subaru']
cars.sort()
print(cars)

cars = ['bmw', 'audi', 'toyota', 'subaru']
print(sorted(cars))
print(cars)
```

- 列表长度：**len()** 函数返回列表的长度。

```python
cars = ['bmw', 'audi', 'toyota', 'subaru']
print(len(cars))
```
