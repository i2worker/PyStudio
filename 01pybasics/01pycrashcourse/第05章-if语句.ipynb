### 5.1 布尔表达式

- 布尔表达式是条件表达式的别名，其结果为真(True)或假(False)。
- 比较运算符：==(相等)、!=(不相等)、<(小于)、>(大于)、<=(小于等于)、>=(大于等于)、in(包含)、not in(不包含)
- 多个条件：and(与)、or(或)

```python
cars = ['audi', 'bmw', 'subaru', 'toyota']
for car in cars:
    if car.lower() == 'bmw':
        print(car.upper())
    else:
        print(car.title())

users = ['andrew', 'carolina', 'david']
user = 'marie'
if user not in users:
    print(f'{user.title()}, you can post a response if you wish.')
```

### 5.2 if语句

- if - elif - else：依次检查每个条件，直到找到匹配项后执行相应代码块，并跳过余下的条件。
- 检查空值：数值0、空值None、空字符串''或""、空列表[]、空字典{}、空元组()都算作空值，返回False。

```python
age = 12
if age < 4:
    price = 0
elif age < 18:
    price = 25
elif age < 65:
    price = 40
elif age >= 65:
    price = 20
print("Your admission cost is $" + str(price) + ".")
```

```python
age = 35
if age < 2:
    time = '婴儿'
elif age < 4:
    time = '幼儿'
elif age < 13:
    time = '儿童'
elif age < 18:
    time = '少年'
elif age < 65:
    time = '中青年人'
elif age >= 65:
    time = '老年人'
print(time)
```

```python
numbers = list(range(1, 10))
for number in numbers:
    if number == 1:
        print('1st')
    elif number == 2:
        print('2nd')
    elif number == 3:
        print('3rd')
    else:
        print(f'{number}th')
```
